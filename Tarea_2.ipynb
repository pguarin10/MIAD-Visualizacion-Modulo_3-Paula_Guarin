### Variable ciudad

```python
"""
Validar variable de ciudad
"""
df['ciudad_limpia'] = df['ciudad'].astype(str).str.strip().str.title()
```

```python
df['ciudad_estandar'] = df['ciudad_limpia']
```

```python
#Washington
df.loc[
    df['ciudad_limpia'].str.contains(
        'Washington',
        case=False,
        na=False
    ),
    'ciudad_estandar'
] = 'Washington'

# NY
df.loc[
    df['ciudad_limpia'].str.contains(
        'New York|Ny ',
        case=False,
        na=False
    ),
    'ciudad_estandar'
] = 'New York'
```

```python
df.loc[
    df['ciudad_limpia'].str.contains(
        'Remote|Work From Home|Worldwide|Rural|Declined|N/A|Area',
        case=False,
        na=False
    ),
    'ciudad_estandar'
] = 'No especificada'
```

```python
UMBRAL = 100
df['ciudad_final'] = df['ciudad_estandar'].where(
    df['ciudad_estandar'].map(df['ciudad_estandar'].value_counts()) >= UMBRAL,
    'Otras ciudades'
)
```

### Otras variables de la base de datos

```python
#Genero
df['genero'].value_counts(dropna=False).head(20)
```

```python
# 1. Crear versión limpia
df['genero_limpio'] = (
    df['genero']
    .astype(str)
    .str.strip()
    .str.title()
)

# 2. Crear género estandar
df['genero_estandar'] = 'Otro / No especificado'

# Mujer
df.loc[df['genero_limpio'] == 'Woman', 'genero_estandar'] = 'Mujer'

# Hombre
df.loc[df['genero_limpio'] == 'Man', 'genero_estandar'] = 'Hombre'

# No binario
df.loc[df['genero_limpio'] == 'Non-Binary', 'genero_estandar'] = 'No binario'

# Preferir no responder (incluye variantes)
df.loc[
    df['genero_limpio'].str.contains(
        'Prefer', case=False, na=False
    ),
    'genero_estandar'
] = 'Otro / No especificado'

# NaN explícitos
df.loc[df['genero_limpio'] == 'Nan', 'genero_estandar'] = 'Otro / No especificado'

# 3. Validación final
df['genero_estandar'].value_counts()
```

```python
#Educación
df['nivel_educativo'].value_counts(dropna=False).head(20)
```

```python
# 1. Crear versión limpia
df['educacion_limpia'] = (
    df['nivel_educativo']
    .astype(str)
    .str.strip()
    .str.title()
)

# 2. Crear variable estándar con valor por defecto
df['educacion_estandar'] = 'Otro / No especificado'

# Secundaria
df.loc[
    df['educacion_limpia'].str.contains('High School', na=False),
    'educacion_estandar'
] = 'Secundaria'

# Educación técnica / incompleta
df.loc[
    df['educacion_limpia'].str.contains('Some College', na=False),
    'educacion_estandar'
] = 'Educación técnica / incompleta'

# Pregrado
df.loc[
    df['educacion_limpia'].str.contains('College Degree', na=False),
    'educacion_estandar'
] = 'Pregrado'

# Maestría
df.loc[
    df['educacion_limpia'].str.contains('Master', na=False),
    'educacion_estandar'
] = 'Maestría'

# Doctorado
df.loc[
    df['educacion_limpia'].str.contains('Phd', case=False, na=False),
    'educacion_estandar'
] = 'Doctorado'

# Grado profesional (MD, JD, etc.)
df.loc[
    df['educacion_limpia'].str.contains('Professional Degree', na=False),
    'educacion_estandar'
] = 'Grado profesional'
df['educacion_estandar'].value_counts()
```
